* **Justificativa:** Manter variáveis numericamente redundantes infla artificialmente o peso desses atributos no modelo e prejudica algoritmos que dependem da independência entre as variáveis. A remoção garante um modelo mais leve e estatisticamente confiável.


#### Teste Qui-Quadrado ($\chi^2$)

```python
import telecom_x_analytics as tax

tabela_decisao = tax.rank_categoricas_completo(df)
display(tabela_decisao)
```

* **Observação:** Esta tabela consolida o cruzamento dos dados antes da fase de treinamento. Note que, além da exclusão das numéricas citadas anteriormente, houve a retirada estratégica de variáveis categóricas redundantes.
* **Por que retirar categóricas específicas?** 1. **Evitar a Armadilha da Variável Dummy (Dummy Variable Trap):** Para variáveis binárias (como Gênero ou Sim/Não), uma coluna é capaz de representar as duas informações. Manter ambas causa redundância perfeita.
    2. **Redução de Dimensionalidade:** Ao remover categorias que não trazem informação nova ou que são constantes, reduzimos o ruído para os modelos de árvore (como XGBoost e Random Forest), permitindo que eles foquem apenas nos atributos que realmente explicam a variação do Churn.
* **Resultado:** O modelo recebe uma base de dados otimizada, o que acelera o treinamento e melhora a interpretabilidade dos resultados.

### Remoção das Colunas para Split

# ➗ Split (Divisão de Treino e Teste)

### 1. Separaçao de X e y

```python
# 1. Primeiro: Separa o que é pergunta (X) do que é resposta (y)
X = df_clean.drop(columns=['Churn'])
y = df_clean['Churn']
```

```python
X.head()
```

```python
y.head()
```

```python
y.value_counts(normalize=True) * 100
```

### 2. Split

```python
# 2. Segundo: Divide em Treino e Teste (O Split propriamente dito)
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y, 
    stratify=y,
    test_size=0.2, 
    random_state=42)

print("Separação e Split concluídos com sucesso!")
```

# 🛠️ Pré Processamento

```python
from sklearn.compose import ColumnTransformer, make_column_selector
from sklearn.preprocessing import StandardScaler, OneHotEncoder

# Criando o pré-processador com FILTROS AUTOMÁTICOS
pre_processamento = ColumnTransformer(
    transformers=[
        # Filtra automaticamente colunas numéricas (float64, int64)
        ('num', StandardScaler(), make_column_selector(dtype_include=['number'])),
        
        # Filtra automaticamente colunas categóricas (object, category)
        ('cat', OneHotEncoder(drop='first', handle_unknown='ignore'), make_column_selector(dtype_exclude=['number']))
    ],
    verbose_feature_names_out=False
)

# Agora é só aplicar nos dados que vieram do Split!
X_train_transformed = pre_processamento.fit_transform(X_train)
X_test_transformed = pre_processamento.transform(X_test)
```

### 🔢 Avaliação Pré Processamento

```python
import numpy as np


print(f"Colunas antes: {X_train.shape[1]}")
print(f"Colunas depois: {X_train_transformed.shape[1]}")


print(f"Média das colunas: {np.mean(X_train_transformed):.2f}") 
# Esperado: 0.00 (ou algo como 1.23e-16, que é quase zero)

print(f"Desvio padrão médio: {np.std(X_train_transformed):.2f}")
# Esperado: Próximo de 1.00b


print(f"Valor mínimo: {X_train_transformed.min():.2f}")
print(f"Valor máximo: {X_train_transformed.max():.2f}")
```

### 🔍 Dataset obtido para Treinamento

```python
# 1. Pegar os nomes das colunas que o pré-processador gerou
# (O método get_feature_names_out é a chave mágica aqui)
nomes_colunas = pre_processamento.get_feature_names_out()

# 2. Criar o DataFrame usando esses nomes
df_treinado_visual = pd.DataFrame(X_train_transformed, columns=nomes_colunas)

# 3. Ver o resultado
df_treinado_visual.head()
```

# 🎯 Treinamento 

### Modelo ⬆️ - Regressão Logística

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report

# 1. Treinamento dos Modelos
# Sem ajuste
log_simples = LogisticRegression(max_iter=1000, random_state=42)
log_simples.fit(X_train_transformed, y_train)

# Com ajuste de balanceamento
log_bal = LogisticRegression(max_iter=1000, class_weight='balanced', random_state=42)
log_bal.fit(X_train_transformed, y_train)

# 2. Gerar e Armazenar as Previsões (A Dica: salvar para reutilizar)
y_pred_log_simples = log_simples.predict(X_test_transformed)
y_pred_log_bal = log_bal.predict(X_test_transformed)
```

### 🔢 Avaliação Performance - Regressão Logística

```python
import telecom_x_analytics as tax


tax.display_side_by_side(y_test, y_pred_log_simples, y_pred_log_bal, "Regressão Logística")
```

* **Observação:** Avaliamos as métricas de Precision, Recall e F1-Score para o modelo base. Esta tabela serve como referência inicial para entender o potencial de previsibilidade linear dos dados da TelecomX.

### 📉 Matriz Confusão - Regressão Logística

```python
import telecom_x_analytics as tax

fig = tax.plot_comparacao_matrizes(y_test, y_pred_log_simples, y_pred_log_bal, "Regressão Logística")

fig.show()
```

* **Observação:** O modelo Balanceado é preferível para a TelecomX porque, embora gere mais 'alarmes falsos' (268 vs 100), ele reduz drasticamente o número de clientes que sairiam sem que percebêssemos (cai de 168 para apenas 81 falhas).

### 📉 Gráfico Coeficientes - Regressão Logística

```python
import telecom_x_analytics as tax

# Para o modelo balanceado
fig_coef = tax.plot_coeficientes_regressao(log_bal, pre_processamento)
fig_coef.show()
```

* **Observação:** Este gráfico detalha o peso e a direção (positiva ou negativa) que cada variável exerce na probabilidade de Churn. É fundamental para validar estatisticamente os insights da análise exploratória.

### 📌 Validação Cruzada - Regressão Logística

```python
from sklearn.model_selection import cross_val_score

# Testando a acurácia em 5 'dobras' diferentes dos dados
scores = cross_val_score(log_bal, X_train_transformed, y_train, cv=5, scoring='accuracy')

print(f"--- VALIDAÇÃO CRUZADA (5-FOLD) ---")
print(f"Acurácia Média: {scores.mean():.2%}")
print(f"Desvio Padrão: {scores.std():.2%}")

if scores.std() < 0.05:
    print("✅ Modelo Robusto: A variação entre os testes é baixa.")
else:
    print("⚠️ Modelo Instável: A performance muda muito dependendo dos dados.")
```

### Modelo 2️⃣ - Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report

# 1. Treinamento dos Modelos de Floresta Aleatória
# Sem ajuste
rf_simples = RandomForestClassifier(n_estimators=100, random_state=42)
rf_simples.fit(X_train_transformed, y_train)

# Com ajuste de balanceamento
rf_bal = RandomForestClassifier(n_estimators=100, class_weight='balanced', random_state=42)
rf_bal.fit(X_train_transformed, y_train)

# 2. Gerar e Armazenar as Previsões
y_pred_rf_simples = rf_simples.predict(X_test_transformed)
y_pred_rf_bal = rf_bal.predict(X_test_transformed)
```

### 🔢 Avaliação Performance - Random Forest

```python
import telecom_x_analytics as tax


tax.display_side_by_side(y_test, y_pred_rf_simples, y_pred_rf_bal, "Random Forest")
```

* **Observação:** Verificamos a performance do modelo de florestas aleatórias. Esperamos uma evolução nas métricas em relação ao modelo linear ao capturar relações mais complexas entre os serviços.

### 📉 Matriz Confusão - Random Forest

```python
import telecom_x_analytics as tax

fig = tax.plot_comparacao_matrizes(y_test, y_pred_rf_simples, y_pred_rf_bal, "Random Forest")
fig.show()
```

* **Observação:** Diferente da Regressão Logística, a Random Forest mostrou-se insensível ao balanceamento de classes, mantendo um alto número de Falsos Negativos (210). Para o objetivo de salvar o máximo de clientes, este modelo é inferior à Regressão Logística Balanceada (que falhou em apenas 81 casos).

### 📉 Gráfico Feature Importance - Random Forest

```python
import telecom_x_analytics as tax

fig = tax.plot_feat_importance_rf(rf_simples, nomes_colunas)
fig.show()
```

* **Observação:** O modelo destaca quais características foram mais decisivas para as "árvores de decisão". Variáveis como Tenure e Mensalidade costumam ganhar destaque nesta fase.

### Modelo 3️⃣ - XGBoost

```python
# !pip install xgboost
```

```python
# Convertendo o alvo para numérico (0 e 1)
y_train_xgb = y_train.map({'No': 0, 'Yes': 1})
y_test_xgb = y_test.map({'No': 0, 'Yes': 1})
```

```python
from xgboost import XGBClassifier
from sklearn.metrics import classification_report, confusion_matrix

# Criando o modelo
xgb_model = XGBClassifier(
    n_estimators=100,
    max_depth=4,
    learning_rate=0.1,
    scale_pos_weight=3, # Balanceamento das classes
    random_state=42,
    eval_metric='logloss'
)

# Treinando
xgb_model.fit(X_train_transformed, y_train_xgb)

# Previsões
y_pred_xgb = xgb_model.predict(X_test_transformed)
```

### 🔢 Avaliação Performance - XGBoost

```python
import telecom_x_analytics as tax

tax.display_side_by_side_xgb(y_test_xgb, y_pred_xgb, "XGBoost")
```

* **Observação:** Análise detalhada da performance do algoritmo de Gradient Boosting. Este modelo é conhecido por sua alta eficiência em problemas de classificação binária desbalanceados.

### 📈 Matriz Confusão - XGBoost

```python
import telecom_x_analytics as tax

fig = tax.plot_matriz_single_xgboost(y_test_xgb, y_pred_xgb, "XGBoost")
fig.show()
```

* **Observação:** O modelo XGBoost consolidou-se como a melhor solução para o problema de Churn da TelecomX. Enquanto a Random Forest apresentou dificuldades em lidar com o desbalanceamento das classes, o XGBoost, aliado à técnica de SMOTE (ou balanceamento), conseguiu capturar 296 dos 369 casos reais de Churn no conjunto de teste.

### 📌 Score AUC - XGBoost

```python
from sklearn.metrics import roc_auc_score

# Calculando a probabilidade (necessário para o AUC)
y_proba_xgb = xgb_model.predict_proba(X_test_transformed)[:, 1]
auc_xgb = roc_auc_score(y_test_xgb, y_proba_xgb)

print(f"O AUC do XGBoost é: {auc_xgb:.4f}")
```

### 📈 Gráfico Feature Importance - XGBoost

```python
import telecom_x_analytics as tax

fig = tax.plot_feat_importance_xg(xgb_model, nomes_colunas)
fig.show()
```

* **Observação:** Diferente dos outros modelos, o XGBoost 'entendeu' que a fidelização contratual é muito mais forte que o valor da mensalidade para segurar o cliente na base.

## 1. Sumário Executivo
Este documento apresenta os resultados da pipeline de dados desenvolvida para a **TelecomX**, focada na antecipação do cancelamento de clientes `Churn`. O objetivo é transformar dados históricos em ações preventivas para garantir a sustentabilidade da receita.

* **Diagnóstico:** Identificamos que **26,4%** da nossa base de clientes cancelou o serviço recentemente. Este volume representa uma perda significativa de receita recorrente e justifica a implementação de uma estratégia de retenção baseada em dados.

### Plano de Ação TelecomX:
1.  **Programa de Estabilização de Contrato:** Migrar clientes "mês a mês" para planos anuais através de benefícios progressivos.
2.  **Otimização Financeira:** Incentivar o uso de métodos de pagamento automático para reduzir o atrito do pagamento manual.
3.  **Auditoria de Experiência em Fibra Óptica:** Investigar a relação entre faturas altas e percepção de qualidade para reduzir o `Churn` de ticket alto.
4.  **Campanha de Valor para Seniores:** Desenvolver pacotes de serviços de suporte e segurança digital focados no público sênior sem dependentes.
5.  **Sistema de Alerta Antecipado (Early Warning):** Integrar o modelo XGBoost ao CRM para sinalizar clientes de "Alto Risco" assim que atingirem o 3º mês de contrato mensal.

## 5. Conclusão Geral e Evolução do Projeto
Este relatório consolida a **Parte 2** de um estudo aprofundado sobre a base de clientes da TelecomX. Enquanto a **Parte 1** consistiu em uma análise exploratória (EDA) abrangente para entender o comportamento geral da base, esta **segunda etapa** focou na maturidade preditiva do projeto.

Para alcançar os resultados aqui apresentados, realizamos um refinamento crítico do dataset, eliminando variáveis redundantes ou de baixo impacto identificadas na fase inicial. Esse processo de "limpeza estratégica" permitiu um conjunto de dados mais equilibrado e focado, facilitando o treinamento de modelos mais robustos e precisos. O diagnóstico final é claro: o `Churn` está concentrado em serviços de alto valor e contratos de baixa fidelidade. Com esse plano de ação proposto, a TelecomX está agora equipada não apenas com dados, mas com uma ferramenta de inteligência capaz de proteger sua receita e aumentar a longevidade da relação com seus clientes.
